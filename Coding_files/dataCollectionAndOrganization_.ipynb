```python
def merge_files(path,top,numberOfFiles):

    """
        Merging reqired numbre of CSV files top either from top or bottom
    """
    
    df_lst = []

    for file in os.listdir(path):
        df_lst.append(path+'/'+file)

    if top==True:
        df_lst = df_lst[:numberOfFiles]   
    else:
        df_lst = df_lst[-numberOfFiles:]

    
    for i in df_lst:
        file = pd.read_csv(i)
        print(i , file.describe())

    df = pd.concat(map(pd.read_csv,[i for i in df_lst]))

    return df
```

```python
df_training=merge_files("/content/drive/MyDrive/Gas_Turbine/Dataset",top=True,numberOfFiles=3)
```

```python
df_testing=merge_files(path="/content/drive/MyDrive/Gas_Turbine/Dataset",top=False,numberOfFiles=2)
```

```python
def save_file(df,name,path,index=False):

    """
        Saving csv files to a location
    """
    
    location = path+'/'+name+'.csv'
    df.to_csv(location,index=index)
    print("File saved")
```

```python
save_file(df_training,name="training_data",path="/content/drive/MyDrive/Gas_Turbine/Dataset")
```

```python
save_file(df_testing,name="testing_data",path="/content/drive/MyDrive/Gas_Turbine/Dataset")
```
